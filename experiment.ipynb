# Agent Inverse Solver Experiment
tests the LLM agent's ability to find the thermal diffusivity $\alpha$ across various simulation times ($t_{final}$).

```python
import os
import asyncio
import pandas as pd
from mcp_client import run_agent

import time
import traceback
```

### Run the Experiment
Iterating across $t_{final} \in \{1, 5, 10\}$ with 5 attempts each.

*Note: The MCP Server dynamically picks up the `HEAT_T_FINAL` environment variable on each loop.*

```python
t = 1.0

os.environ["HEAT_T_FINAL"] = str(t)
print(f"\nTesting t_final = {t}...")
its = []
    
for attempt in range(1, 11):
    try:
        iterations = await run_agent()
        print(f"  Attempt {attempt} completed in {iterations} iterations.")
        its.append(iterations)

    except Exception as e:
        traceback.print_exc() 
        print(f"  Attempt {attempt} failed: {e}")
        
    time.sleep(5)  
```
